# Sprawdzian 1: Wczytywanie plików
To notebook przygotowujący do pierwszego sprawdzianu z pracy na plikach tekstowych.

## Wstęp teoretyczny
Pliki tekstowe możemy w Pythonie wczytywać **linijka po linijce**. Każda linijka kończy się specjalnym znakiem nazywanym *znakiem nowej linii* (\n).\nJest to niewidoczny znak, który informuje komputer, że kolejny fragment tekstu powinien trafić do nowego wiersza.\nPodczas wczytywania pliku można zdecydować, czy chcemy pobrać wszystkie linie od razu, czy może przetwarzać je kolejno, np. w pętli.

## Przygotowanie przykładowego pliku
Na potrzeby ćwiczeń zapiszemy prosty plik tekstowy z kilkoma linijkami, aby mieć na czym pracować.

```python
przyklad = 'pierwsza linia\ndruga linia\ntrzecia linia'
with open('przyklad.txt', 'w', encoding='utf-8') as plik:
    plik.write(przyklad)
print('Plik przyklad.txt został zapisany.')
```

## Wczytanie wszystkich linijek naraz
Funkcja `readlines()` zwraca listę, w której każda pozycja to kolejna linia z pliku łącznie ze znakiem nowej linii.
Możemy też użyć `read()` i `splitlines()`, aby pozbyć się znaków nowej linii.

```python
with open('przyklad.txt', 'r', encoding='utf-8') as plik:
    wszystkie_linie = plik.readlines()
print(wszystkie_linie)

with open('przyklad.txt', 'r', encoding='utf-8') as plik:
    wszystkie_linie_bez_znaku = plik.read().splitlines()
print(wszystkie_linie_bez_znaku)
```

## Wczytywanie linijka po linijce w pętli `for`
Przetwarzając plik w pętli, Python sam zwraca kolejne linie. Dzięki temu możemy analizować lub modyfikować dane na bieżąco i nie musimy trzymać całego pliku w pamięci.

```python
with open('przyklad.txt', 'r', encoding='utf-8') as plik:
    for numer, linia w enumerate(plik, start=1):
        linia = linia.rstrip('
')  # usuwamy znak nowej linii z końca
        print(f'Linia {numer}: {linia}')
```
