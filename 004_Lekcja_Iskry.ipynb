## Wstęp teoretyczny

Własną funkcję tworzymy słowem kluczowym `def`, podając nazwę i listę parametrów w nawiasie. W środku funkcji zapisujemy instrukcje, które mają się wykonać, a wynik możemy zwrócić przez `return`. Funkcję wywołujemy po nazwie, przekazując argumenty pasujące do parametrów.

### Przykład: prosta funkcja dodająca liczby

Poniżej jest krótki przykład funkcji, która dodaje dwie liczby, oraz jej wywołania.


```python
def dodaj(a, b):
    return a + b

wynik = dodaj(3, 5)
print(wynik)
```

## Rekurencja w skrócie

Rekurencja to sytuacja, gdy funkcja wywołuje samą siebie. Kluczowe są **warunek stopu** (aby nie wywoływać się w nieskończoność) i **krok rekurencyjny**, który upraszcza problem.

### Przykład: rekurencyjne dodawanie dwóch liczb dodatnich

Funkcja `add_positive(a, b)` dodaje dwie nieujemne liczby, zmniejszając `b` aż do zera.


```python
def add_positive(a, b):
    if b == 0:
        return a
    return add_positive(a + 1, b - 1)

print(add_positive(4, 3))
```
